```python
import pandas as pd
import re
import string

from collections import Counter
```

```python
df = pd.read_csv('pubg_review.csv')
df
```

Preprocessing

```python
def preprocessing(text):
    text = text.lower()
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'https\S+', '', text)
    text = re.sub(r'@\w+|#\w+', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'\s{2,}', ' ', text)
    return text.strip()

df['cleaned'] = df['content'].apply(preprocessing)
df
```

```python
def sentiment(score):
    if score >= 4:
        return "Positive"
    elif score == 3:
        return "Neutral"
    else:
        return "Negative"
    
df['sentiment'] = df['score'].apply(sentiment)
df
```

```python
keywords_aspect = {
    'grafik': ['grafik', 'visual', 'hd', 'gambar'],
    'bug': ['bug', 'crash', 'error', 'hang', 'macet'],
    'gameplay': ['gameplay', 'mode', 'fitur', 'battle', 'rank'],
    'cheater': ['cheat', 'cheater', 'curang', 'hacker'],
    'kontrol': ['kontrol', 'tombol', 'control'],
    'update': ['update', 'pembaruan', 'patch', 'upgrade'],
    'server': ['lag', 'server', 'ping', 'delay', 'network', 'lancar'],
}

def aspect(text):
    terdeteksi = []
    for aspek, keywords in keywords_aspect.items():
        for keyword in keywords:
            if keyword in text:
                terdeteksi.append(aspek)
                break
    return terdeteksi if terdeteksi else ['Lainnya']

df['aspect'] = df['cleaned'].apply(aspect)
df

df.to_csv('pubg_review_cleaned.csv', index=False)
```

```python
aspek_sentimen = []

for _, row in df.iterrows():
    for aspek in row['aspect']:
        aspek_sentimen.append((aspek, row['sentiment']))

# Buat DataFrame ringkasan
summary = pd.DataFrame(aspek_sentimen, columns=['aspect', 'sentimen'])
pivot = summary.pivot_table(index='aspect', columns='sentimen', aggfunc='size', fill_value=0)

print(pivot)
```
