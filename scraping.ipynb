```python
from google_play_scraper import reviews, Sort
import pandas as pd
```

```python
package = 'com.tencent.ig'

result, _ = reviews(
    package,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=1000,
)

df = pd.DataFrame(result)

df.to_csv('pubg_review.csv', index=False)

df
```
